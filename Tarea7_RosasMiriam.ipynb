```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

Los **datos** son los siguientes:

```python
def f(x):
    return np.exp(-x**2/2)/np.sqrt(2*np.pi)
```

```python
x=np.linspace(-3,3)
y=f(x)
plt.figure(figsize=(8,6))
plt.plot(x,y,'*b',label='datos')
plt.legend(loc='best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

Definimos la función de **grado 1**:

```python
def obj1(b,x,y):
    return np.sum((y-np.array([x**i for i in range(2)]).T.dot(b))**2)
b0=np.random.random((2,))
res=opt.minimize(obj1, b0, args=(x,y))
yhat1=np.array([x**i for i in range(2)]).T.dot(res.x)
plt.figure(figsize=(8,6))
plt.plot(x,y,'*b',label='datos')
plt.plot(x,yhat1,'r',label='ajuste 1')
plt.legend(loc='best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
def obj2(b,x,y):
    return np.sum((y-np.array([x**i for i in range(3)]).T.dot(b))**2)
b0=np.random.random((3,))
res=opt.minimize(obj2,b0,args=(x,y))
yhat2=np.array([x**i for i in range(3)]).T.dot(res.x)
plt.figure(figsize=(8,6))
plt.plot(x,y,'*b',label='datos')
plt.plot(x,yhat1,'r',label='ajuste 1')
plt.plot(x,yhat2,'g',label='ajuste 2')
plt.legend(loc='best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

Claculemos el **error** para cada grado de lo polinomios que ajustamos previamente (es decir, del 1 al 7):

```python
e_ms=[]
def obj(b,x,y,n):
    return np.sum((y-np.array([x**i for i in range(n+1)]).T.dot(b))**2)
for i in range(7):
    b0=np.random.random((i+2,))
    res=opt.minimize(obj,b0,args=(x,y,i+1))
    yhat=np.array([x**j for j in range(i+2)]).T.dot(res.x)
    e_ms.append(sum((y-yhat)**2))
plt.figure(figsize=(8,5))
plt.plot(np.arange(7)+1,e_ms,'o')
plt.xlabel('orden')
plt.ylabel('error')
plt.show()
```

Podemos darnos cuenta de que el error del grado 1 es de casi 1, lo que significa que casi el 100% de las veces el modelo se equivoca. En cambio, el error para los grados 2 y 3 es menor, pues aproximadamente el 20% de las veces no acierta. Los polinomios de grado 4,5,6 y 7 tienen aproximadamente un error de 0.05, así que para optimizar el modelo y no tener un polinomio de grado muy alto, elejimos el **polinomio de grado 4**. A continuación realizamos el ajusto con ridge de la función que elegimos:

```python
def obj_ridge(b,x,y,n,l):
    return np.sum((y-np.array([x**i for i in range(n+1)]).T.dot(b))**2)+l*np.linalg.norm(b)**2
b0=np.random.random((5,))
res=opt.minimize(obj_ridge,b0,args=(x,y,4,0.001))
yhat4_ridge=np.array([x**j for j in range(5)]).T.dot(res.x)
plt.figure(figsize=(8,6))
plt.plot(x,y,'*b',label ='datos')
plt.plot(x,yhat4,'c',label='ajuste 4')
plt.plot(x,yhat4_ridge,'r',label='ajuste 4 con ridge')
plt.legend(loc='best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

Finalmente, comparamos el error entre la función de **grado 4** que habíamos hecho previamente y su respectivo **ajuste con el método ridge**. Podemos ver con la gráfica que realmente no son muy diferentes, por lo que la diferencia del error entra ambas es **insignificante**. Considero que no vale la pena calcularlo. Cualquiera de los dos modelos (tanto con ajuste ridge como sin él), son buenos.
